### DATA COLLECTION AND PROCESSING 

```python
sonar_data = pd.read_csv("sonar data.csv" , header = None ) # SINCE THERE IS NO COLUMN NAME ON EACH(header = None) 
sonar_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

### separating data and Labels


```python
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]

print(X)
print(Y)
```

### Training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

-- stratify=Y: This is a crucial parameter when dealing with imbalanced datasets. 

-- It ensures that the test set and the training set have roughly the same proportion of each class as the original dataset. 

```python
print (X.shape , Y.shape , X_train.shape , X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

### Model Training using Logisitic regression

```python
# load logistic regression model

# model = LogisticRegression()
```

### Training the logistic regression model with training data

```python
# model.fit ( X_train , Y_train )
from sklearn.model_selection import GridSearchCV  #  GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100], 'max_iter': [100, 200, 300]}
grid = GridSearchCV(LogisticRegression(), param_grid, cv=5)
grid.fit(X_train, Y_train)

# Get the best model
print("Best Parameters:", grid.best_params_)
model = grid.best_estimator_ 
```

```python
# Training the best model
model.fit(X_train, Y_train)

```

	•	C=10 means the model performs best when the regularization strength (C) is set to 10.


### Model Evaluation 

```python
# accuracy on training data 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on text data : ', test_data_accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(Y_test, y_pred))

print("\nClassification Report:")
print(classification_report(Y_test, y_pred))
```

### Build a predictive sys

```python
import random
```

```python
# Randomly select a row index
random_index = random.randint(0, len(sonar_data) - 1)

print(f"Selected row index: {random_index}")

# Extract the row (including all columns)
selected_row = sonar_data.iloc[random_index]

print("Selected Row Data:")  # Print the full row
print(selected_row)
```

```python
input_data = selected_row.iloc[:-1].values  # Excludes last column (label)

# Changing the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the np array as we are predicting for one instance
input_data_res = input_data_as_numpy_array.reshape(1, -1)

# Predict using the trained model
prediction = model.predict(input_data_res)

if prediction[0] == 'R':
    print("Prediction: Rock")
else:
    print("Prediction: Mine")
```
